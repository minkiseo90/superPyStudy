### 159 오늘의 요일 계산하기(localtime)

```python
from time import localtime

weekdays = ['월요일', '화요일', '수요일', '목요일', '금요일', '토요일', '일요일']

t = localtime()
# print(t)
today = '%d-%d-%d' %(t.tm_year, t.tm_mon, t.tm_mday)

week = weekdays[t.tm_wday]
print('[%s]는 [%s]입니다'%(today,week))
```

### 160 프로그램 실행 시간 계산하기 (datetime.now)

### 165 스택 구현하기(append, pop)

```python
mystack = []

def putdata(data):
    global mystack
    mystack.append(data)
    
def popdata():
    global mystack
    if len(mystack) == 0:
        return None
    return mystack.pop()

putdata('data1')
putdata([3,4,5,6])
putdata(12345)

print('<스택 상태>: ', end = ""); print(mystack)

ret = popdata()
while ret != None:
    print('스택에서 데이터 추출 : ',end=""); print(ret)
    print('<스택 상태> : ',end = ""); print(mystack)
    ret = popdata()
```

### 166 문장에 나타나는 문자 빈도수 계산하기

```python
def getTextFreq(filename):
    with open(filename,'r') as f:
        text = f.read()
        fa={}
        for c in text:
            if c in fa:
                fa[c] += 1
            else:
                fa[c] = 1
    return fa

ret = getTextFreq('mydata.txt')
ret = sorted(ret.items(), key = lambda x:x[1], reverse = True)

for c, freq in ret:
    if c == '\n' or c == ' ':
        continue
    print('[%c] -> [%d]회 나타남' %(c, freq))
```

### 168 파일에서 특정 단어 개수 계산하기

```python
def countWord(filename, word):
    with open(filename, 'r') as f:
        text = f.read()
        text = text.lower()
        pos = text.find(word)
        count = 0
        while pos != -1:
            count += 1
            pos = text.find(word, pos+1)
    return count

word = input('mydata.txt 에서 개수를 구할 단어를 입력하세요:')
word = word.lower()
ret = countWord('mydata.txt',word)
print('[%s]의 개수 : %d' %(word, ret))
```

### 173 인터넷에 있는 대용량 파일을 내 PC로 저장하기

```python
from urllib.request import urlopen

BUFSIZE = 256*1024

fileurl = 'https://www.python.org/ftp/python/3.5.2/python-3.5.2.exe'
filename = fileurl.split('/')[-1]

try:
    with urlopen(fileurl) as f:
        with open(filename,'wb') as h:
            buf = f.read(BUFSIZE)
            while buf:
                h.write(buf)
                buf = f.read(BUFSIZE)
except Exception as e:
    print(e)
```

### 174 10MB 파일을 1MB 파일 10개로 분리하기

```python
filename = 'python-3.5.2.exe'
subsize = 1024*1024*3
suffix = 0

with open(filename,'rb') as f:
    buf = f.read(subsize)
    while buf:
        subfilename = filename +'_'+str(suffix)
        with open(subfilename,'wb') as h:
            h.write(buf)
            print('[%s] 완료]'%(subfilename))
        buf = f.read(subsize)
        suffix +=1
```

### 175 1MB 파일 10개를 합쳐서 10MB 파일로 만들기

```python
BUFSIZE = 256*1024
merge_filename = 'ret.exe'
filelist = ['python-3.5.2.exe_'+str(x) for x in range(10)]

with open(merge_filename,'wb') as f:
    for filename in filelist:
        print('[%s]합치는중 : ' %filename)
        with open(filename, 'rb') as h:
            buf = h.read(BUFSIZE)
            while buf:
                f.write(buf)
                buf = h.read(BUFSIZE)
print('파일합치기 완료')
```

### 176 파일을 ZIP 압축파일로 만들기

```python
from zipfile import *

def compressZip(zipname, filename):
    print('[%s]를 [%s]로 압축' %(filename, zipname))
    with ZipFile(zipname, 'w') as ziph:
        ziph.write(filename)
        
    print('압축완료')
    
filename = 'mydata.txt'
zipname = filename +'.zip'
compressZip(zipname, filename)
```

### 179 로또 번호 추출기 만들기

```python
from random import shuffle
from time import sleep

gamenum = input('로또게임횟수는?')

for i in range(int(gamenum)):
    balls = [x+1 for x in range(45)]
    ret = []
    for j in range(6):
        shuffle(balls)
        number = balls.pop()
        ret.append(number)
    ret.sort()
    print('로또번호[%d]:'%(i+1),end = "")
    print(ret)
    sleep(1)
```
